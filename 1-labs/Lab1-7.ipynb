````python
int_types = [np.int8, np.int16, np.int32, np.int64]
print("Integer types:")
for dtype in int_types:
    info = np.iinfo(dtype)
    print(f"{dtype.__name__}: min={info.min}, max={info.max}")

uint_types = [np.uint8, np.uint16, np.uint32, np.uint64]
print("\nUnsigned integer types:")
for dtype in uint_types:
    info = np.iinfo(dtype)
    print(f"{dtype.__name__}: min={info.min}, max={info.max}")

float_types = [np.float16, np.float32, np.float64]
print("\nFloating-point types:")
for dtype in float_types:
    info = np.finfo(dtype)
    print(f"{dtype.__name__}: min={info.min}, max={info.max}, eps={info.eps}")
````

9. Как отсортировать массив по n-му столбцу?

````python
array = np.array([
    [5, 2, 9],
    [1, 7, 6],
    [8, 3, 4],
    [2, 6, 1]
])

# индекс столбца, по которому нужно сортировать
n = 1

# индексы строк
sorted_indices = np.argsort(array[:, n])

sorted_array = array[sorted_indices]

print("Original")
print(array)

print("\nSorted array by column", n, ":")
print(sorted_array)
````

3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
import numpy as np
def split_blood_pressure(bp):
    systolic, diastolic = map(int, bp.split('/'))
    return pd.Series([systolic, diastolic])

df[['Systolic Pressure', 'Diastolic Pressure']] = df['Blood Pressure (s/d)'].apply(split_blood_pressure)

conditions = [
    (df['Stress Levels'] <= 4),
    (df['Stress Levels'] > 4) & (df['Stress Levels'] <= 7),
    (df['Stress Levels'] > 7)
]
choices = ['low', 'medium', 'high']
df['stress_category'] = np.select(conditions, choices)

average_blood_pressure = df.groupby('stress_category').agg(
    Average_Systolic=('Systolic Pressure', 'mean'),
    Average_Diastolic=('Diastolic Pressure', 'mean')
).reset_index()

print(average_blood_pressure) #Неправда
````

````python
Ответ: Неправда, более высокий уровень стресса не говорит о более высоком кровяном давлении
````

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

````python
import numpy as np

df['Chronic Diseases'] = df['Chronic Diseases'].map({'Yes': 1, 'No': 0})

percentiles = np.percentile(df['BMI'], [30, 60, 75])
percentile_30, percentile_60, percentile_75 = percentiles

group_high_bmi = df[df['BMI'] > percentile_75]
group_medium_bmi = df[(df['BMI'] >= percentile_30) & (df['BMI'] <= percentile_60)]

chronic_high_bmi = group_high_bmi['Chronic Diseases'].mean()
chronic_medium_bmi = group_medium_bmi['Chronic Diseases'].mean()

if chronic_high_bmi > chronic_medium_bmi:
    print("У людей с ИМТ выше 75-го перцентиля хронические заболевания встречаются чаще, чем у людей с ИМТ в пределах 30-60 перцентиля.")
elif chronic_high_bmi < chronic_medium_bmi:
    print("У людей с ИМТ в пределах 30-60 перцентиля хронические заболевания встречаются чаще, чем у людей с ИМТ выше 75-го перцентиля.")
else:
    print("Частота хронических заболеваний одинакова в обеих группах.")
````

9. Правда ли, что люди с высоким уровнем дохода имеют больший уровень физической активности, чем люди с низким доходом

````python
activity_counts = df.groupby(['Income Level', 'Physical Activity Level']).size().unstack().fillna(0)


total_counts = activity_counts.sum(axis=1)

activity_percentages = activity_counts.div(total_counts, axis=0) * 100

print("Доля людей с каждым уровнем физической активности по уровням дохода:")
print(activity_percentages)

high_income_activity = activity_percentages.loc['High']
low_income_activity = activity_percentages.loc['Low']

if high_income_activity['High'] > low_income_activity['High']:
    print("\nЛюди с высоким уровнем дохода имеют больший уровень физической активности по сравнению с людьми с низким доходом.")
else:
    print("\nЛюди с высоким уровнем дохода не имеют большего уровня физической активности по сравнению с людьми с низким доходом.")
````

10. Какая доля мужчин с высокой частотой пребывания на солнце имеет высокий заработок?

````python
men_data = df[df['Gender'] == 'Male']

sun_exposure_75th_percentile = men_data['Sun Exposure'].quantile(0.75)

print(f"Порог для высокой частоты пребывания на солнце (75-й перцентиль): {sun_exposure_75th_percentile:.2f} часов в неделю")

high_sun_exposure_men = men_data[men_data['Sun Exposure'] > sun_exposure_75th_percentile]

high_income_count = high_sun_exposure_men[high_sun_exposure_men['Income Level'] == 'High'].shape[0]
total_high_sun_exposure_count = high_sun_exposure_men.shape[0]

if total_high_sun_exposure_count > 0:
    high_income_ratio = high_income_count / total_high_sun_exposure_count * 100
else:
    high_income_ratio = 0

print(f"Доля мужчин с высокой частотой пребывания на солнце, которые имеют высокий уровень дохода: {high_income_ratio:.2f}%")
````

5. Постройте круговую диаграмму отображающую как ментальное здоровье связано с паттернами сна

````python
def categorize_mental_health(status):
    if status <= 40:
        return 'Низкое'
    elif status <= 70:
        return 'Среднее'
    else:
        return 'Высокое'

df['Mental Health Category'] = df['Mental Health Status'].apply(categorize_mental_health)

categories_count = df.groupby(['Mental Health Category', 'Sleep Patterns']).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
for category in categories_count.index:
    sizes = categories_count.loc[category]
    plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', startangle=140, shadow=True)
    plt.axis('equal')  # Равные оси для круговой диаграммы
    plt.title(f'Паттерны сна в зависимости от ментального здоровья: {category}')
    plt.show()
````
